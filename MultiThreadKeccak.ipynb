```python
import itertools
import threading
import numpy as np
import sys
from Crypto.Hash import keccak
from datetime import datetime
import time

# Alphabet initialization

lower_case = "abcdefghijklmnopqrstuvwxyz"
upper_case = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "0123456789"
symbols = "_"

Merged = lower_case + upper_case + numbers + symbols


# Main parameters

Input = "(first_param, second_param)"
Desired_number_of_zeros = 5
STR = "0" * Desired_number_of_zeros


########################################################    Functions   ################################################

def keccak_calculation(Word):
    from Crypto.Hash import keccak
    import binascii

    For_hash = bytes(Word, 'utf-8')
    keccak = keccak.new(data=For_hash, digest_bits=256).digest()
    return binascii.hexlify(keccak)


def string_update(Word):
    new_str = str()
    for i in str(Word):
        if i in Merged:
            new_str += i
        else:
            continue
    return new_str

def thread_job(Merged, L, Input): #L1, L2):
    
    got_it = 0
    global stop_threads
    
    Input_str = str(Input)
    L_arr = [L] #[L1, L2]
    
    for k in range(1):
        L = L_arr[k]
        if got_it == 1:
            break
        # print("Current length of testing words is:" , L)
        permutations = itertools.permutations(Merged, L)
        for subset in permutations:

            if stop_threads:
                # print('End of the main program in Thread:', L)
                break

            current_word = string_update(subset)+Input_str

            Hash = keccak_calculation(current_word)
            str_of_decoded_hash = str(Hash.decode('utf-8'))
            is_in_Hash = str_of_decoded_hash.find(STR, 0, Desired_number_of_zeros)

            if is_in_Hash != -1:
                print("Bingo!", str_of_decoded_hash)
                print("Word:", current_word)
                print("Thread number:", L)
                # print("".join(i))
                got_it = 1
                print("Required time is:", datetime.now() - start_time)
                stop_threads = True
                break
            elif is_in_Hash == -1:
                continue

    #  sys.stdout.flush()

def run_threads(count, Merged, Input):
    # A1 = np.arange(30)
    # A2 = np.arange(30,60)[::-1]
    #
    # for i in range(30):
    #     threads = threading.Thread(target=thread_job, args=(Merged, A1[i], A2[i]))

    threads = [
        threading.Thread(target=thread_job, args=(Merged, 1, Input)),
        threading.Thread(target=thread_job, args=(Merged, 2, Input)),
        threading.Thread(target=thread_job, args=(Merged, 3, Input)),
        threading.Thread(target=thread_job, args=(Merged, 4, Input)),
        threading.Thread(target=thread_job, args=(Merged, 5, Input)),
        threading.Thread(target=thread_job, args=(Merged, 6, Input)),
        threading.Thread(target=thread_job, args=(Merged, 7, Input)),
        threading.Thread(target=thread_job, args=(Merged, 8, Input)),
        threading.Thread(target=thread_job, args=(Merged, 9, Input)),
        threading.Thread(target=thread_job, args=(Merged, 10, Input)),
        threading.Thread(target=thread_job, args=(Merged, 11, Input)),
        threading.Thread(target=thread_job, args=(Merged, 12, Input)),
        threading.Thread(target=thread_job, args=(Merged, 13, Input)),
        threading.Thread(target=thread_job, args=(Merged, 14, Input)),
        threading.Thread(target=thread_job, args=(Merged, 15, Input)),
        threading.Thread(target=thread_job, args=(Merged, 16, Input)),
        threading.Thread(target=thread_job, args=(Merged, 17, Input)),
        threading.Thread(target=thread_job, args=(Merged, 18, Input)),
        threading.Thread(target=thread_job, args=(Merged, 19, Input)),
        threading.Thread(target=thread_job, args=(Merged, 20, Input)),
        threading.Thread(target=thread_job, args=(Merged, 21, Input)),
        threading.Thread(target=thread_job, args=(Merged, 22, Input)),
        threading.Thread(target=thread_job, args=(Merged, 23, Input)),
        threading.Thread(target=thread_job, args=(Merged, 24, Input)),
        threading.Thread(target=thread_job, args=(Merged, 25, Input)),
        threading.Thread(target=thread_job, args=(Merged, 26, Input)),
        threading.Thread(target=thread_job, args=(Merged, 27, Input)),
        threading.Thread(target=thread_job, args=(Merged, 28, Input)),
        threading.Thread(target=thread_job, args=(Merged, 29, Input)),
        threading.Thread(target=thread_job, args=(Merged, 30, Input)),
        threading.Thread(target=thread_job, args=(Merged, 31, Input)),
        threading.Thread(target=thread_job, args=(Merged, 32, Input)),
        threading.Thread(target=thread_job, args=(Merged, 33, Input)),
        threading.Thread(target=thread_job, args=(Merged, 34, Input)),
        threading.Thread(target=thread_job, args=(Merged, 35, Input)),
        threading.Thread(target=thread_job, args=(Merged, 36, Input)),
        threading.Thread(target=thread_job, args=(Merged, 37, Input)),
        threading.Thread(target=thread_job, args=(Merged, 38, Input)),
        threading.Thread(target=thread_job, args=(Merged, 39, Input)),
        threading.Thread(target=thread_job, args=(Merged, 40, Input)),
        threading.Thread(target=thread_job, args=(Merged, 41, Input)),
        threading.Thread(target=thread_job, args=(Merged, 42, Input)),
        threading.Thread(target=thread_job, args=(Merged, 43, Input)),
        threading.Thread(target=thread_job, args=(Merged, 44, Input)),
        threading.Thread(target=thread_job, args=(Merged, 45, Input)),
        threading.Thread(target=thread_job, args=(Merged, 46, Input)),
        threading.Thread(target=thread_job, args=(Merged, 47, Input)),
        threading.Thread(target=thread_job, args=(Merged, 48, Input)),
        threading.Thread(target=thread_job, args=(Merged, 49, Input)),
        threading.Thread(target=thread_job, args=(Merged, 50, Input)),
        threading.Thread(target=thread_job, args=(Merged, 51, Input)),
        threading.Thread(target=thread_job, args=(Merged, 52, Input)),
        threading.Thread(target=thread_job, args=(Merged, 53, Input)),
        threading.Thread(target=thread_job, args=(Merged, 54, Input)),
        threading.Thread(target=thread_job, args=(Merged, 55, Input)),
        threading.Thread(target=thread_job, args=(Merged, 56, Input)),
        threading.Thread(target=thread_job, args=(Merged, 57, Input)),
        threading.Thread(target=thread_job, args=(Merged, 58, Input)),
        threading.Thread(target=thread_job, args=(Merged, 59, Input)),
        threading.Thread(target=thread_job, args=(Merged, 60, Input))
    ]

    for thread in threads:
        thread.start()

    for thread in threads:
        thread.join()

        
########################################################    Main program   ################################################

stop_threads = False
start_time = datetime.now()
print("Threads are starting!")

run_threads(60, Merged, Input)

print('finish all')
```
